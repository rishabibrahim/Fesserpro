```python
# rock paper scissor
hand1=input("hand1")
hand2=input("hand2")
# tie, hand1 win , hand2 win
if hand1 == hand2:
    print("TIE")
elif (hand1=="p" and hand2=="r") or (hand1=="s" and hand2=="p") or (hand1=="r" and hand2=="s"):
    print("hand1 wins")
else:
    print("hand2 wins")
```
